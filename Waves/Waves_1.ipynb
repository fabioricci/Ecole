```python
import matplotlib.pyplot as plt
import numpy as np
import math as mh
from math import sin, pi
import random

A=round(random.uniform(1.5,4.5),1) #meters
w=round(random.uniform(0.5,2.5),1) #Hz
phi=round(random.uniform(-0.7,0.7),1) #radiants

print('A = ', A, 'm ; w = ', w, 'Hz ; phi = ', phi, 'rad')

t = np.arange(0, 10.1, 0.01)
s = A*np.sin(w*t+phi)
fig, ax = plt.subplots()
plt.subplots_adjust(top=1., bottom=0., left=0.0, right=0.95, hspace=0.25,
                    wspace=0.5)
ax.plot(t,s,'r-')
plt.axhline(y=0.0, color='blue', linestyle='--')
ax.set_xlim(xmin=0)
ax.set_xlim(xmax=10)
ax.set_ylim(ymin=-A-0.5)
ax.set_ylim(ymax=A+0.5)
ax.set_xticks(np.arange(0., 10.1, step=1),minor=False)
ax.set_yticks(np.arange(-A-0.5, A+0.5, 0.5),minor=False)
ax.set_ylabel('x (m)')
ax.set_xlabel('t (s)')
plt.grid()
plt.show()
```
